```python
from glob import glob
import pandas as pd
df_list=[]
for filename in glob("../data/EGRE_DATOS_ABIERTOS*.csv"):
    temp=pd.read_csv(filename,delimiter=";",encoding="latin-1")
    temp.rename({"PERTENENCIA_ESTABLECIMIENTO_SALU":"PERTENENCIA_ESTABLECIMIENTO_SALUD"},inplace=True)
    df_list.append(temp)

df_final = pd.concat(df_list)
df_final["MRUN"] = range(0,df_final.shape[0])

print(df_final.head(3))
print(df_final.tail(3))

```

```python
print(df_final["SEXO"].value_counts())
#print(df_final.describe())
df_final.groupby("SEXO").agg({"GRUPO_EDAD":"nunique"})
```

```python
df_final["DIAS_MAX_DIAS"] = df["GRUPO_EDAD"] #mapa de dias max segun edad
```

```python
import seaborn as sns
import numpy as np
#kernel density plot
sns.boxplot(y="DIAS_ESTADA",data=df_final[df_final["DIAS_ESTADA"] <= np.percentile(df["DIAS_ESTADA"],99)])

```

```python
threshold = df_final["DIAS_ESTADA"].quantile(0.99)
# Filter natively using pandas and plot (with the closing parenthesis!)
sns.boxplot(
    y="DIAS_ESTADA", 
    data=df_final[df_final["DIAS_ESTADA"] <= threshold]
)
```
